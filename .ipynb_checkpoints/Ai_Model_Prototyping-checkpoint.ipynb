```python
import pandas as pd
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from sklearn.ensemble import IsolationForest
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
import matplotlib.pyplot as plt

# 1. Load simulated data
df = pd.read_csv("data/simulated_filter_consumption.csv", parse_dates=["date"])

# 2. Weekly aggregation
df_w = df.groupby(pd.Grouper(key="date", freq="W"))["filters_used"].sum().reset_index()
train, test = df_w.iloc[:-4].set_index("date"), df_w.iloc[-4:].set_index("date")

# 3. Seasonality decomposition (period=12 weeks)
decomp = seasonal_decompose(train["filters_used"], model="additive", period=12)
decomp.plot()
plt.suptitle("Seasonal Decomposition")
plt.show()

# 4. ARIMA forecasting
model = ARIMA(train["filters_used"], order=(1,1,1), seasonal_order=(1,1,1,12))
fit   = model.fit()
fcst  = fit.forecast(steps=4)

# 5. Evaluate
mae  = mean_absolute_error(test["filters_used"], fcst)
mape = mean_absolute_percentage_error(test["filters_used"], fcst)
print(f"MAE: {mae:.2f}, MAPE: {mape:.2%}")

# 6. Plot actual vs forecast
plt.figure()
plt.plot(train.index, train["filters_used"], label="Train")
plt.plot(test.index,  test["filters_used"],  label="Actual")
plt.plot(test.index,  fcst, label="Forecast", linestyle="--")
plt.legend()
plt.title("Weekly Filters: Actual vs Forecast")
plt.show()

# 7. Anomaly detection
# 7a. Rolling-window
roll_mean = df_w["filters_used"].rolling(4, center=True).mean()
roll_std  = df_w["filters_used"].rolling(4, center=True).std()
df_w["anom_rw"] = (df_w["filters_used"] - roll_mean).abs() > 2*roll_std

# 7b. IsolationForest
iso = IsolationForest(contamination=0.05, random_state=42)
df_w["anom_if"] = iso.fit_predict(df_w[["filters_used"]]) == -1

print("\nRecent anomalies (last 8 weeks):")
print(df_w.tail(8)[["date","filters_used","anom_rw","anom_if"]])
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from sklearn.ensemble import IsolationForest
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

# 1. Load or simulate data
csv_path = "data/simulated_filter_consumption.csv"
if os.path.exists(csv_path):
    df = pd.read_csv(csv_path, parse_dates=["date"])
else:
    # Simulate if CSV not found
    sites = ["NorthLab", "SouthClinic", "EastHospital"]
    devices_per_site = 5
    dates = pd.date_range("2025-01-01", "2025-06-30", freq="D")
    rows = []
    for date in dates:
        for site in sites:
            for i in range(devices_per_site):
                rows.append({"date": date, "site_id": site, "device_id": f"{site[:3].upper()}-{i+1:03d}"})
    df = pd.DataFrame(rows)
    np.random.seed(42)
    df["base_usage"] = np.random.normal(12, 3, size=len(df))
    df["day_of_year"] = df["date"].dt.dayofyear
    df["seasonal_mul"] = 1 + 0.15 * np.sin(2*np.pi*df["day_of_year"]/365)
    df["weekday"] = df["date"].dt.weekday
    df["weekly_mul"] = df["weekday"].apply(lambda x: 0.7 if x>=5 else 1.0)
    spike_days = {site: np.random.choice(dates, 10, replace=False) for site in sites}
    df["spike_mul"] = df.apply(lambda r: np.random.uniform(1.5,2.0) 
                              if r["date"] in spike_days[r["site_id"]] else 1.0, axis=1)
    df["filters_used"] = (df["base_usage"]*df["seasonal_mul"]*
                          df["weekly_mul"]*df["spike_mul"]).round().astype(int)
    os.makedirs(os.path.dirname(csv_path), exist_ok=True)
    df.to_csv(csv_path, index=False)
    print(f"Simulated data saved to {csv_path}")

# 2. Weekly aggregation
df_w = df.groupby(pd.Grouper(key="date", freq="W"))["filters_used"].sum().reset_index()

# 3. Split train/test
train = df_w.iloc[:-4].set_index("date")
test  = df_w.iloc[-4:].set_index("date")

# 4. Determine period for decomposition
period = max(2, len(train) // 2)

# 5. Seasonal decomposition
decomp = seasonal_decompose(train["filters_used"], model="additive", period=period)
decomp.plot()
plt.suptitle("Seasonal Decomposition")
plt.tight_layout()
plt.show()

# 6. ARIMA forecasting
model = ARIMA(train["filters_used"], order=(1,1,1), seasonal_order=(1,1,1,period))
fit = model.fit()
forecast = fit.forecast(steps=4)

# 7. Forecast evaluation
mae  = mean_absolute_error(test["filters_used"], forecast)
mape = mean_absolute_percentage_error(test["filters_used"], forecast)
print(f"Forecast MAE: {mae:.2f}")
print(f"Forecast MAPE: {mape:.2%}\n")

# 8. Plot actual vs forecast
plt.figure()
plt.plot(train.index, train["filters_used"], label="Train")
plt.plot(test.index,  test["filters_used"],  label="Actual")
plt.plot(test.index,  forecast,               label="Forecast", linestyle="--")
plt.legend()
plt.title("Weekly Filters: Actual vs Forecast")
plt.tight_layout()
plt.show()

# 9. Anomaly detection
window = 4
rolling_mean = df_w["filters_used"].rolling(window=window, center=True).mean()
rolling_std  = df_w["filters_used"].rolling(window=window, center=True).std()
df_w["anomaly_rw"] = (df_w["filters_used"] - rolling_mean).abs() > 2*rolling_std

iso = IsolationForest(contamination=0.05, random_state=42)
df_w["anomaly_if"] = iso.fit_predict(df_w[["filters_used"]]) == -1

# 10. Display anomalies
print("Recent Anomalies (Last 10 Weeks):")
print(df_w.tail(10)[["date","filters_used","anomaly_rw","anomaly_if"]].to_string(index=False))
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima.model import ARIMA
from sklearn.ensemble import IsolationForest
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

# 1. Load or simulate data (same as before)
csv_path = "data/simulated_filter_consumption.csv"
if os.path.exists(csv_path):
    df = pd.read_csv(csv_path, parse_dates=["date"])
else:
    # simulate same as earlier
    # (omitted here for brevity; use your existing simulate block)
    pass

# 2. Weekly aggregation and enforce a WSUN freq
df_w = (
    df
    .groupby(pd.Grouper(key="date", freq="W"))["filters_used"]
    .sum()
    .to_frame()
    .reset_index()
)
df_w = df_w.set_index("date").asfreq("W")  # now index.freq == "W-SUN"

# 3. Split into train/test
train = df_w.iloc[:-4]
test  = df_w.iloc[-4:]

# 4. Choose a modest seasonal period (e.g. 8 weeks) if you have 16 points
seasonal_period = 8 if len(train) >= 16 else None

# 5. If we can do a seasonal decomposition, do it; otherwise skip
if seasonal_period:
    try:
        decomp = seasonal_decompose(train["filters_used"], model="additive", period=seasonal_period)
        decomp.plot()
        plt.suptitle(f"Seasonal Decomposition (period={seasonal_period})")
        plt.tight_layout()
        plt.show()
    except ValueError:
        print(f"Skipping seasonal_decompose: insufficient data for period={seasonal_period}")

# 6. Build ARIMA model; if seasonal_period is None or too few points, drop seasonal component
if seasonal_period and len(train) >= 2 * seasonal_period:
    order = (1, 1, 1)
    seasonal_order = (1, 1, 1, seasonal_period)
else:
    order = (1, 1, 1)
    seasonal_order = (0, 0, 0, 0)

model = ARIMA(
    train["filters_used"],
    order=order,
    seasonal_order=seasonal_order
)
fit = model.fit()
forecast = fit.forecast(steps=len(test))

# 7. Evaluate forecast
mae  = mean_absolute_error(test["filters_used"], forecast)
mape = mean_absolute_percentage_error(test["filters_used"], forecast)
print(f"Forecast MAE: {mae:.2f}")
print(f"Forecast MAPE: {mape:.2%}\n")

# 8. Plot actual vs forecast
plt.figure()
plt.plot(train.index, train["filters_used"], label="Train")
plt.plot(test.index,  test["filters_used"],  label="Actual")
plt.plot(test.index,  forecast,               label="Forecast", linestyle="--")
plt.legend()
plt.title("Weekly Filters: Actual vs Forecast")
plt.tight_layout()
plt.show()

# 9. Anomaly detection (unchanged)
window = 4
roll_mean = df_w["filters_used"].rolling(window, center=True).mean()
roll_std  = df_w["filters_used"].rolling(window, center=True).std()
df_w["anom_rw"] = (df_w["filters_used"] - roll_mean).abs() > 2 * roll_std
iso = IsolationForest(contamination=0.05, random_state=42)
df_w["anom_if"] = iso.fit_predict(df_w[["filters_used"]]) == -1

# 10. Print recent anomalies
print("Recent Anomalies (last 10 weeks):")
print(df_w.tail(1000)[["filters_used","anom_rw","anom_if"]])

#  after youve plotted train/test/forecast 
plt.tight_layout()
plt.savefig("forecast_chart.png", dpi=150)
plt.close()
```
